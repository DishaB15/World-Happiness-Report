```python
import pandas as pd
import seaborn as sns
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2,constrained_layout=True,figsize=(12,8))

sns.barplot(x='GDP per capita',y='Country or region',data=data2019.nlargest(10,'GDP per capita'),ax=axes[0,0],palette="Blues_d")

sns.barplot(x='Social support' ,y='Country or region',data=data2019.nlargest(10,'Social support'),ax=axes[0,1],palette="YlGn")

sns.barplot(x='Healthy life expectancy' ,y='Country or region',data=data2019.nlargest(10,'Healthy life expectancy'),ax=axes[1,0],palette='OrRd')

sns.barplot(x='Freedom to make life choices' ,y='Country or region',data=data2019.nlargest(10,'Freedom to make life choices'),ax=axes[1,1],palette='YlOrBr')
```

```python
data2017.rename(columns = {'Happiness.Score' : 'Score', 'Economy..GDP.per.Capita.' :'GDP per capita','Family' :'Social support',
                           'Health..Life.Expectancy.' : 'Healthy life expectancy'
                          
                          }, inplace = True)
```

```python
d= data2019[(data2019['Country or region'].isin(['Finland','Canada','South Sudan','United States']))]
```

```python
ax = d.plot(y="Social support", x="Country or region", kind="bar",color='C8')
d.plot(y="GDP per capita", x="Country or region", kind="bar", ax=ax, color="C9")
d.plot(y="Healthy life expectancy", x="Country or region", kind="bar", ax=ax, color="C6")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.kdeplot(data2017['Healthy life expectancy'],color='red', Label='2017')
sns.kdeplot(data2018['Healthy life expectancy'],color='blue', Label='2018')
sns.kdeplot(data2019['Healthy life expectancy'],color='pink', Label='2019')
plt.title('Health and life expectancy over the Years',size=20)
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.kdeplot(data2017['Score'],color='green', Label='2017')
sns.kdeplot(data2018['Score'],color='blue', Label='2018')
sns.kdeplot(data2019['Score'],color='orange', Label='2019')
plt.title('Happiness Score over the Years',size=20)
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.kdeplot(data2017['GDP per capita'],color='red', Label='2017')
sns.kdeplot(data2018['GDP per capita'],color='blue', Label='2018')
sns.kdeplot(data2019['GDP per capita'],color='pink', Label='2019')
plt.title('GDP over the years',size=20)
plt.show()
```
